**Question 4. Create a loyalty flag for existing customers using the transform() and loc() funtions.**

```python
# Apply the transform function on the "order_number" column to generate the maximum orders for each user.

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

**Question 6. Create a "spending flag" based on the avg price across each customers orders.**

```python
# Start by creating a new 'avg_price' column based on user_id

ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check descriptive statistics on the prices column for each type of customer.

ords_prods_merge.groupby('loyalty_flag').agg({'avg_price' : ['mean','min','max']})
```

**Question 7. Create an order frequency flag that marks the regularity of a customers ordering according to the average "days_since_prior_order".**

```python
# Create a new column to show a customers average length in between orders. Group by 'user_id'

ords_prods_merge['avg_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(100)
```

```python
# Create an order frequency flag that marks the regularity of a user's spending behavior using the loc() function 

ords_prods_merge.loc[ords_prods_merge['avg_days_between_orders'] > 20, 'frequency_flag'] = 'Non-Frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['avg_days_between_orders'] > 10) & (ords_prods_merge['avg_days_between_orders'] <= 20), 'frequency_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_days_between_orders'] <= 10, 'frequency_flag'] = "Frequent Customer"
```

```python
# Check the frequency of each value

ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
