# If-Statements with User-Defined Functions

```python
# Define a function 

def price_label(row):
    
    if row['prices']<= 5:
        return 'low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'mid range product'
    elif (row['prices'] > 15):
        return 'high range product'
    else: return np.nan
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Use the max function to check what the most expensive product within the subset is

df['prices'].max()
```

# If Statements with For-Loops

For the Instacart dataframe, begin by creating a new column in your ords_prods_merged dataframe that summarizes how busy each day of the week is. Start by printing the "orders_day_of_the_week" column.

```python
# Check the frequency of values in the "orders_day_of_week" column

df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Use a for-loop statement to create a new column denoting the "busiest day", "least busy", and "regularly busy" days.

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
    if value == 0:
        result.append("busiest day")
    elif value == 4:
        result.append("least busy")
    else:
        result.append("regularly busy")
```

```python
# Combine the above list with the df_ords_prods_merged dataframe

df_ords_prods_merged['busiest_day'] = result
```

**Question 2. Create a new column that shows the two busiest days and the two slowest days using a suitable method.** 

```python
# Use if-statements with for-loops to create new column

result_2 = []

for value in df_ords_prods_merged["orders_day_of_week"]:
    if value == 0 or value == 1:
        result_2.append("busiest days")
    elif value == 4 or value == 3:
        result_2.append("slowest days")
    else:
        result_2.append("regularly busy")
```

```python
list_quantile = list(df['order_hour_of_day'].value_counts(dropna = False).quantile([0.25,0.75]))
```

```python
dict_hour_counts = df['order_hour_of_day'].value_counts(dropna = False).to_dict()
```

```python
result_t = []

for value in df['order_hour_of_day']:
    for key, val in dict_hour_counts.items():
        if value == key:
            if val < list_quantile[0]:
                result_t.append("fewest orders")
            elif val >= list_quantile[0] and val < list_quantile[1]:
                result_t.append("average orders")
            elif val >= list_quantile[1]:
                result_t.append("most orders")
```

```python
df['busiest_hours'] = result_t
```
