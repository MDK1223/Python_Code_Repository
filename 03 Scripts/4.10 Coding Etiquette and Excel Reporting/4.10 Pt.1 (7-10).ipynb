**7. Aggregate max, mean, and min variables on a customer-profile level for usage and frequency expenditure.**

```python
df_final.shape
```

```python
# Income Group Aggregat

income_group_agg = df_final.groupby('income_group').agg({'days_since_prior_order': ['min', 'mean', 'max'], 'max_order': ['min', 'mean', 'max'], 'avg_price': ['min', 'mean', 'max']})
```

```python
income_group_agg
```

```python
# Age Group Aggregate

age_group_agg = df_final.groupby('age_group').agg({'days_since_prior_order': ['min', 'mean', 'max'], 'max_order': ['min', 'mean', 'max'], 'avg_price': ['min', 'mean', 'max']})
```

```python
age_group_agg
```

```python
# Family Status Aggregate

fam_stat_agg = df_final.groupby('fam_status').agg({'days_since_prior_order': ['min', 'mean', 'max'], 'max_order': ['min', 'mean', 'max'], 'avg_price': ['min', 'mean', 'max']})
```

**9. Produce suitable charts to illustrate your results and export them in your project folder.**

```python
# Income Group By Region

income_group_region_bar = income_group_region.plot.bar(color = ['red', 'green', 'gold', 'blue'])
plt.title('Regional  Group sizes')
plt.xlabel('Total')
plt.ylabel('Income Group')
plt.legend('Region')
plt.legend(loc = 'center left', bbox_to_anchor = (1.0,0.5))
```

```python
# Age Group by Region

age_group_region_bar = age_group_region.plot.bar(color = ['red', 'green', 'gold', 'blue'])
plt.title('Regional Age Group Sizes')
plt.ylabel('Total')
plt.xlabel('Age Group')
plt.legend('Region')
plt.legend(loc = 'center left', bbox_to_anchor = (1.0, 0.5))
```

```python
# Fam Status by Region

fam_stat_region_bar = fam_stat_region.plot.bar(color = ['red', 'green', 'gold', 'blue'])
plt.title('Family Status by Region')
plt.xlabel('Family Status')
plt.ylabel('Total')
plt.legend().set_title('')
```

```python
# Family Size by Region

fam_size_region_bar = fam_size_region.plot.bar(color = ['red', 'green', 'gold', 'blue'])
plt.title('Family Size by Region')
plt.xlabel('Family Size')
plt.ylabel('Total')
plt.legend(loc = 'center left', bbox_to_anchor=(1.0, 0.5))
```

```python
# Income Group by Department

income_group_dept_bar = income_group_dept.plot.barh(stacked = True, color = ['red', 'green', 'gold'])
plt.title('Department by Income Group')
plt.xlabel('Frequency')
plt.ylabel('Department')
plt.legend().set_title('')
```

```python
# Age Group by Department

age_group_dept_bar = age_group_dept.plot.barh(stacked = True, color = ['red', 'green', 'gold', 'blue'])
plt.title('Department by Age Group')
plt.xlabel('Frequency')
plt.ylabel('Department')
plt.legend().set_title('')
```

```python
# Fam Stat by Dept

fam_stat_dept_bar = fam_stat_dept.plot.barh(stacked = True, color = ['red', 'green', 'gold', 'blue'])
plt.title('Department by Family Status')
plt.xlabel('Frequency')
plt.ylabel('Department')
plt.legend().set_title('')
```

```python
# Fam Size by Dept

fam_size_dept_bar = fam_size_dept.plot.barh(stacked = True, color = ['red', 'green', 'gold', 'blue'])
plt.title('Department by Family Size')
plt.xlabel('Frequency')
plt.ylabel('Department')
plt.legend().set_title('')
```
